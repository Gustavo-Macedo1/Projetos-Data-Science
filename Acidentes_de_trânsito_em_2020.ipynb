# Pré-processamento

## Tratando as datas com datetime

Queremos apenas saber o mês, dado que a planilha é sobre o ano, e o dia do mês pouco importa, dado que o dia da semana está explicitado em outra coluna

Necessário usar:
- pd.to_datetime
- dt.strftime("%d-%m-%Y")

```python
# 1. Converte tudo para datetime
dados.data_inversa = pd.to_datetime(dados.data_inversa)

# 2. Muda o formato de toda a coluna
dados.data_inversa = dados.data_inversa.dt.strftime("%m")

dados.data_inversa.info()
```

```python
datas = {
    '01' : 'Janeiro',
    '02' : 'Fevereiro',
    '03' : 'Março',
    '04' : 'Abril',
    '05' : 'Maio',
    '06' : 'Junho',
    '07' : 'Julho',
    '08' : 'Agosto',
    '09' : 'Setembro',
    '10' : 'Outubro',
    '11' : 'Novembro',
    '12' : 'Dezembro'
}
```

```python
dados.data_inversa = dados.data_inversa.map(datas)
dados.data_inversa.head()
```

Por fim, devemos fazer a troca do nome da coluna usando a função **rename**

```python
dados = dados.rename(columns={'data_inversa' : 'mes'})
dados.mes.head()
```

## Análise simples

### Qual a causa de acidentes mais frequente?

```python
dados.causa_acidente.unique()
```

```python
mais_frequente = dados.causa_acidente.value_counts()
mais_frequente.head(5)
```

Portanto, os maiores motivo dos acidentes de trânsito no ano de 2020 foram:

- Falta de atenção
- Desobediência das normas de trânsito
- Velocidade incompativel
- Ingestão de álcool
- Defeito mecânico no veículo

```python
dados.causa_acidente.info()
```

Assim, temos 63530 dados não nulos nessa coluna. Vamos pegar as frequências relativas dos primeiros 5 maiores motivos de acidentes.


```python
contagem = dados.causa_acidente.count()
f1 = (22410 / contagem) * 100
f2 = (8802 / contagem) * 100
f3 = (5883 / contagem) * 100
f4 = (5069 / contagem) * 100
f5 = (3742 / contagem) * 100

print(f'Frequência relativa - Falta de atenção: {f1:.2f}%')
print(f'Frequência relativa - Desobediência das normas de trânsito: {f2:.2f}%')
print(f'Frequência relativa - Velocidade incompatível: {f3:.2f}%')
print(f'Frequência relativa - Ingestão de álcool: {f4:.2f}%')
print(f'Frequência relativa - Defeitos mecânicos no veículo: {f5:.2f}%')
```

```python
soma_maiores = f1 + f2 + f3 + f4 + f5
print(f'A soma das 5 maiores causas para acidentes resultam em {soma_maiores:.2f}% dos acidentes de 2020.')
```

```python
causas_diferentes = len(dados.causa_acidente.unique())
causas_diferentes
```

Assim, 3.5 a cada 10 acidentes são causados por falta de atenção, evidenciando a importância de se reforçar políticas públicas que diminuam o uso do celular e outras distrações durante a condução do veículo.

Além disso, mesmo que tenhamos 58 tipos diferentes de causas, as 5 maiores concentram 72.26% dos acidentes ocorridos em 2020, mostrando que as políticas públicas devem, primariamente, conter acidentes causados por tais fatores.
