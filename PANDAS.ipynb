```python
#Qual o número total de casos?
total_cases = covid_df.new_cases.sum()
print(f'O número total de casos é {int(total_cases)}.')
```

```python
#Qual a taxa de mortes?
death_ratio = covid_df.new_deaths.sum() / total_cases
print(f'A taxa de mortes é de {death_ratio*100:.2f}%.')
```

```python
#Qual o total de testes? Sabe-se que 935310 testes não foram contabilizados na tabela
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
print(f"O total de testes realizados no período foi de: {total_tests:.0f}.")
```

```python
#Qual a taxa de testes positivos?
positive_ratio = total_cases / total_tests
print(f'A taxa de casos positivos foi de {positive_ratio*100:.2f}% dos testes.')
```

```python
#Qual o total absoluto de casos negativos?
negative_total = total_tests - total_cases
print(f'O número absoluto de testes negativos foi {negative_total:.0f} no período analisado.')
```

```python
# Gerando Series para casos > 1000 : gera um Series com Bools
high_cases = covid_df.new_cases > 1000
```

```python
#Gerando sub-frame para casos > 1000: gera um data-frame filho
high_cases_df = covid_df[high_cases] #Vai no DF original e pega só as linhas em que os casos são > 1000, ou seja, os True da Series
```

```python
#Em quais dias a taxa de positivos foi maior que a média?
#Média = positive_ratio
high_ratio = covid_df.new_cases / covid_df.new_tests > positive_ratio
high_ratio_df = covid_df[high_ratio]
high_ratio_df
```

```python
#Se operarmos duas colunas, criamos um novo Series com os valores da operação
dif_teste_caso = covid_df.new_tests - covid_df.new_cases
dif_teste_caso
```

```python
#Podemos adicionar a coluna no DF original usando o método padrão de dicionário
covid_df['dif_tc'] = dif_teste_caso
covid_df
```

```python
#Podemos também retirar colunas do DF usando a função drop
covid_df.drop(columns=['dif_tc'], inplace=True) #inplace em False faz retornar uma cópia do data frame, mas sem a coluna
covid_df
```

```python
#Podemos ordenar as linhas de acordo com a(s) coluna(s) desejada(s) com a função sort_values
#Aqui vamos ordenar para pegar as 10 linhas com os maiores números de novos casos
#Essa função RETORNA uma subDF do DF original
covid_df.sort_values('new_cases', ascending=True).head(10)
```

```python
#Vemos que tem o valor -148 na tabela, podendo ser um erro.
#Dependendo do caso, deletamos a linha ou trocamos por 0, média geral ou média entre os valores adjacentes.
#Nesse caso, como o DF está ordenado por data, vamos pegar a média dos dois valores adjacentes.
#Usamos a função .at para acessar a linha. Não usamos .loc pois ela gera uma copia do DF original.

covid_df.at[172, 'new_cases'] = (covid_df.at[171, 'new_cases'] + covid_df.at[173, 'new_cases']) / 2
covid_df.at[172, 'new_cases']
```

```python
#Colocando os valores de data do formato 'object' para o formato 'datetime'
covid_df.date = pd.to_datetime(covid_df.date)
covid_df.info()
```

```python
#Caso precise alterar o tipo de uma coluna para o valor numérico, usamos o método to_numeric(). Exemplo:

#covid_df.date = pd.to_numeric(covid.df.date, errors='coerce')

#O parâmetro 'errors' recebe um valor 'coerce', que indica que valores não numéricos (strings, por exemplo), se tornem NaNs na coluna.
```

```python
#Criando colunas adicionais separadas por tempo, usando dados datetime e a classe pd.DatetimeIndex()

covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
covid_df
```

# Cumulative Sum
Vai somando linha a linha na coluna desejada

```python
covid_df['total_cases'] = covid_df['new_cases'].cumsum()
```

```python
covid_df['total_tests'] = covid_df['new_tests'].cumsum()
```

```python
covid_df['total_deaths'] = covid_df['new_deaths'].cumsum()
```

```python
#Mas e se quisermos mais informações?

url_locations = 'https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv'
urlretrieve(url_locations, 'locations.csv')
```

```python
#Checando o arquivo
os.listdir()
```

```python
#Lendo o arquivo
locations_df = pd.read_csv('locations.csv')
locations_df
```

```python
italy = locations_df[locations_df.location == 'Italy'] #Linha de interesse
italy
```

```python
#Queremos fazer um merge com essa linha, como fazer?
#Primeiro devemos ter uma coluna em comum entre os dois dataframes de interesse.

covid_df['location'] = 'Italy'
covid_df
```

```python
#Depois, podemos fazer o merge com a função dataframe_orig.merge(dataframe, on='coluna_comum')
merge_df = covid_df.merge(italy, on='location')
merge_df
```

```python
#Agora com as novas informações podemos calcular novas métricas

merge_df['cases_per_million'] = merge_df.new_cases * 1e6 / merge_df.population
merge_df['deaths_per_million'] = merge_df.new_deaths * 1e6 / merge_df.population
merge_df['tests_per_million'] = merge_df.new_tests * 1e6 / merge_df.population

merge_df
```

```python
#Agora vamos exportar o arquivo para um novo arquivo CSV
#Para fazê-lo, vamos primeiro criar o dataframe resultante apenas com as informações necessárias

result_df = merge_df[['new_cases',
                     'total_cases',
                     'new_deaths',
                     'total_deaths',
                     'new_tests',
                     'total_tests',
                     'cases_per_million',
                     'deaths_per_million',
                     'tests_per_million']]
result_df
```

```python
#Agora usamos o método .to_csv('nome_do_arquivo.csv', index=None) O index tira essa coluna do 0 até o 247 do arquivo csv.
#Lembrando que quando abre, ele já vem com esse índice padrão.

result_df.to_csv('results.csv', index=None)
```
