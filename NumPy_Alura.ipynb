#Análise numérica e regressão linear

Vamos tentar achar uma reta que resuma bem o crescimento das maçãs

```python
plt.plot(meses, cidade1)
```

```python
x = meses
y = 0.8*x + 80 #chutando valores
```

```python
plt.plot(meses, cidade1)
plt.plot(x, y)
```

Vemos que a curva está um pouco acima, mas como quantizar essa diferença? -> Usamos a raiz do quadrado da soma dos desvios (sqrt(soma_desvios)**2)

```python
desvios = cidade1 - y
desvios
```

```python
desvios = np.power(desvios,2)
```

```python
soma_desvios = np.sum(desvios)
soma_desvios
```

```python
np.sqrt(soma_desvios) # Agora está quantizada, quanto mais próxima de 0, melhor
```

```python
# NP.COPY() -> função que copia um array de forma desvinculada, se atribuirmos um array a outro, e modificarmos o original, o filho também é modificado!!!
# mas se usarmos essa função, ele é desvinculado, pode alterar de boas
```

```python
np.linalg.norm(cidade1 - y)
```

Agora vamos aprender a fazer a regressão linear sem chutar valores

Primeiro, usamos a seguinte equação para calcular o coeficiente linear:

imagem_2023-05-22_234928153.png

```python
n = np.size(cidade1)
Y = cidade1
X = meses
```

```python
a = (n * np.sum(X*Y) - np.sum(X) * np.sum(Y))/(n * np.sum(X**2) - np.sum(X)**2)
# a = 0.43520212145512926
```

Agora vamos calcular o coeficiente linear (b). Para isso, usaremos a seguinte fórmula:

image.png

```python
b = np.mean(Y) - a * np.mean(X)
# b = 79.64892274792834
```

```python
y = a * x + b
```

```python
plt.plot(meses, y)
plt.plot(meses, cidade1)
plt.title('Preço das maçãs em Moscow em 4 anos')
plt.xlabel('Meses')
plt.ylabel('Preço')
plt.plot(47.4, 47.4*a+b, '*r') #estimativa do preço para 47.4 meses
plt.plot(130, 130*a+b, '*r') #estimativa para o preço em 130 meses (previsão)
```

```python
np.linalg.norm(cidade1 - y)
#diferença menor
```

# Análise numérica eficiente e randomização

```python
#Gerando números float aleatórios
coef_aleatorios = np.random.uniform(low=0.4, high=0.9, size=100)
```

```python
norma = np.array([])
for i in range(100):
  norma = np.append(norma, np.linalg.norm(cidade1 - (coef_aleatorios[i]*X + b)))

norma
```

```python
np.min(norma) # Pegando o menor valor da norma -> 101.89950976208759
np.where(norma == 101.89950976208759) # Qual o índice do menor valor? -> (array([21]),)
# Então podemos pegar o coeficiente linear aleatório da posição 21 e usá-lo
menor_a = coef_aleatorios[21] # -> 0.4448743693307077
```

Como gerar os mesmos números aleatórios? Os números na verdade são pseudoaleatórios e dependem de uma semente. Fixando uma semente, garantimos a reprodutibilidade dos dados aleatórios.

```python
np.random.seed(15)
np.random.uniform(30, 100, 50)

#Se rodar o código 2x ele repete os mesmos números
```

Unindo o array de normas com o array de coeficientes aleatorios

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
citricas = np.loadtxt('citrus.csv', delimiter=',', usecols=np.arange(1,5,1), skiprows=1)
```

```python
print(f'Tabela:\n')
print(citricas)
print(f'\nShape: ', end='')
print(np.shape(citricas))
print(f'\nNúmero de dimensões: ', end='')
np.ndim(citricas)
```

```python
diametro_laranjas = citricas[0:5000,0]
peso_laranjas = citricas[0:5000,1]
diametro_toranjas = citricas[5000:,0]
peso_toranjas = citricas[5000:,1]
```

```python
plt.plot(diametro_laranjas, peso_laranjas)
plt.plot(diametro_toranjas, peso_toranjas)
plt.legend(['Laranjas', 'Toranjas'])
```

Fazendo regressão linear com os gráficos em questão

```python
plt.plot(diametro_laranjas, peso_laranjas)
```

```python
X = diametro_laranjas
Y = peso_laranjas
n = np.size(peso_laranjas) #nº de elementos do eixo Y
```

```python
a = (n * np.sum(X*Y) - np.sum(X) * np.sum(Y))/(n * np.sum(X**2) - np.sum(X)**2)
a
```

```python
b = np.mean(Y) - a * np.mean(X)
b
```

```python
plt.plot(diametro_laranjas, peso_laranjas)
plt.plot(X, a*X+b)
```

```python
Y1 = peso_toranjas
X1 = diametro_toranjas
n = np.size(peso_toranjas)
```

```python
a_tor = (n * np.sum(X1*Y1) - np.sum(X1) * np.sum(Y1))/(n * np.sum(X1**2) - np.sum(X1)**2)
a_tor
```

```python
b_tor = np.mean(Y1) - a_tor * np.mean(X1)
b_tor
```

```python
plt.plot(diametro_toranjas, peso_toranjas)
plt.plot(X1, a_tor*X1 + b_tor)
```
